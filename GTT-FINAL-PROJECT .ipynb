```python
from tkinter import *
from tkinter import filedialog
from PIL import ImageTk, Image
import cv2
import pytesseract 
from tkinter import scrolledtext
from tkinter import filedialog,messagebox
from pathlib import Path
import pyttsx3
import glob


pytesseract.pytesseract.tesseract_cmd = 'C:\\Users\\avi\\Anaconda3\\envs\\project\\Library\\bin\\tesseract.exe'
global element
element = 0


root = Tk()
root.title("Text Extraction")
root.state('zoomed')
root.resizable(width = False,height = True)
root.configure(bg = "black")

def home_screen():
    MainFrame = Frame(root, bg = "gray")
    
    MainFrame.pack()
    root_lbl = Label(MainFrame,text = "Text Extraction From Image",bg = "black", fg = "white",font = ('',40,'italic','underline'), width=50, height=2, relief='raised', borderwidth=5)
    root_lbl.pack(fill = X,anchor = N,expand = True)
    
    q_frame = Frame(MainFrame,bg = 'gray', width = 750, height=800, relief='flat', borderwidth=5)
    q_frame.pack(side = LEFT,fill = Y,anchor = W,expand = True)

    a_frame = Frame(MainFrame,bg = 'yellow',width = 800, height=800, relief='flat', borderwidth=5)
    a_frame.pack(side = RIGHT,fill = Y,anchor = E,expand = True)


    a_lbl = Label(a_frame,text="Output Will Be Shown Here", fg = 'black',bg = 'yellow',font = ('',20,'bold','underline'))
    a_lbl.place(relx = 0.25,rely = 0.01)

    br_file = Label(q_frame,text = "Path :",bg = "gray",font = ('',28,'bold'),relief = 'flat',width = 6)
    br_file.place(relx = 0.001,rely = 0.001)


    br_btn = Button(q_frame,text = "Browse",command = lambda : browse_file(src_entr,image_list),bg = "gray",font = ('',18,'bold'),relief = 'raised')
    br_btn.place(relx = 0.7,rely = 0.01)

    src_entr = Entry(q_frame , relief = 'raised',font=('',30,'bold'),width = 15)
    src_entr.place(relx = 0.2,rely = 0.01)
    
    a_lbl = Label(q_frame,text="Image Will Be Shown Here\n Press Next button for first image", fg = 'black',bg = 'gray',font = ('',20,'bold','underline'))
    a_lbl.place(relx = 0.2,rely = 0.4)

    r_btn = Button(q_frame,text = "Extract",command = lambda : extract(q_frame,a_frame,src_entr,image_list,element),width = 15,bg = "gray",font = ('',18,'bold'),relief = 'raised')
    r_btn.place(relx = 0.32,rely = 0.1)
    
    global image_list
    image_list = []
    
def extract(q_frame,a_frame_2,src_entr,image_list,element):
    
    
    voice_btn = Button(a_frame_2,text = "Voice",command = lambda : text_to_speech(texts),width = 15,bg = "green",font = ('',18,'bold'),relief = 'raised')
    voice_btn.place(relx = 0.35,rely = 0.92)
    
    prev_btn = Button(q_frame,text = "Prev",command = lambda : prev_image(q_frame,a_frame_2,element,image_list,src_entr),width = 15,bg = "green",font = ('',18,'bold'),relief = 'raised')
    prev_btn.place(relx = 0.0001,rely = 0.1)
    
    next_btn = Button(q_frame,text = "Next",command = lambda : next_image(q_frame,a_frame_2,element,image_list,src_entr),width = 15,bg = "green",font = ('',18,'bold'),relief = 'raised')
    next_btn.place(relx = 0.64,rely = 0.1)
    
    #print(image_list)
    
    if len(src_entr.get()) < 2:
        messagebox.showwarning("Warning","Please select the path first !!!!")
        return
    
def browse_file(src_entr,img_files):
    file_path=filedialog.askdirectory()
    #filedialog.askopenfilename(filetypes=[("Images", ".jpg .jpeg .png")])
    src_entr.delete(0,END)
    src_entr.insert(0,file_path)
    img_files.clear()
    dirr = Path(src_entr.get())
    ext = ['png', 'jpg', 'gif' , 'jpeg']
    [img_files.extend(glob.glob(f"{dirr}/*" + e)) for e in ext]
    
        
def text_to_speech(texts):
    engine = pyttsx3.init()
    engine.say(texts)
    engine.runAndWait()
    engine.stop()

    
def next_image(q_frame,a_frame_2,element,img_files,src_entr):
    voice_btn = Button(a_frame_2,text = "Voice",command = lambda : text_to_speech(texts),width = 15,bg = "green",font = ('',18,'bold'),relief = 'raised')
    voice_btn.place(relx = 0.35,rely = 0.92)
    
    if(element >= len(img_files)):
        messagebox.showinfo("IMAGES","No more Images !!!!")
        return
    else:
        try:
            #print("n",element)
            def upload(q_frame):
                global photo
                path = img_files[element]
                img_frame = Frame(q_frame,bg = 'gray', width = 748, height=550, relief='flat')
                img_frame.place(relx = 0.000001,rely = 0.2)
                img  = Image.open(path).resize((740,525))
                photo=ImageTk.PhotoImage(img)
                lab=Label(img_frame,image=photo,bg = 'black')
                lab.place(relx = .0001,rely = 0.0001 ,height = 530 )

            upload(q_frame)
            
            global texts

            path = img_files[element]
            #print(path)
            
            Actual_image = cv2.imread(path)
            Sample_img = cv2.resize(Actual_image,(400,350))
            Image_ht,Image_wd,Image_thickness = Sample_img.shape
            Sample_img = cv2.cvtColor(Sample_img,cv2.COLOR_BGR2RGB)
            
            texts = pytesseract.image_to_data(Sample_img) 
            texts = pytesseract.image_to_string(Sample_img)
            
            text_area = scrolledtext.ScrolledText(a_frame_2,
                                                  width = 50,  
                                                  height = 100,
                                                  bg = 'gold',
                                                  font = ("",
                                                          15)) 
            text_area.delete('1.0',END)
            text_area.insert(INSERT,"-------------------------------------------O U T P U T-------------------------------------------------")
            text_area.insert(INSERT,"\n")


            text_area.insert(END,f"{texts}")
            text_area.configure(state ='disabled')
            text_area.place(relx = 0.0001,rely = 0.0001,height = 600,width = 775)
            text_area.focus()
            element = element + 1
            extract(q_frame,a_frame_2,src_entr,image_list,element)  

        except:

            a_lbl = Label(q_frame,text="Image Will Be Shown Here", fg = 'black',bg = 'gray',font = ('',20,'bold','underline'))
            a_lbl.place(relx = 0.2,rely = 0.4)
            messagebox.showerror("Unsupported file","File Not Supported !!!!!")
            return
        
        
def prev_image(q_frame,a_frame_2,element,img_files,src_entr):
    voice_btn = Button(a_frame_2,text = "Voice",command = lambda : text_to_speech(texts),width = 15,bg = "green",font = ('',18,'bold'),relief = 'raised')
    voice_btn.place(relx = 0.35,rely = 0.92)
    if(element <= 0):
        messagebox.showinfo("IMAGES","No more Images !!!!")
        
    else:
        element = element - 1
        try:
            #print("P",element)
            
            def upload(q_frame):
                global photo
                path = img_files[element]
                img_frame = Frame(q_frame,bg = 'gray', width = 748, height=550, relief='flat')
                img_frame.place(relx = 0.000001,rely = 0.2)
                img  = Image.open(path).resize((740,525))
                photo=ImageTk.PhotoImage(img)
                lab=Label(img_frame,image=photo,bg = 'black')
                lab.place(relx = .0001,rely = 0.0001 ,height = 530 )

            upload(q_frame)
            
            global texts

            path = img_files[element]
            #print(path)
            
            Actual_image = cv2.imread(path)
            Sample_img = cv2.resize(Actual_image,(400,350))
            Image_ht,Image_wd,Image_thickness = Sample_img.shape
            Sample_img = cv2.cvtColor(Sample_img,cv2.COLOR_BGR2RGB)
            
            texts = pytesseract.image_to_data(Sample_img) 
            texts = pytesseract.image_to_string(Sample_img)

        
            text_area = scrolledtext.ScrolledText(a_frame_2,
                                                  width = 50,  
                                                  height = 100,
                                                  bg = 'gold',
                                                  font = ("",
                                                          15)) 
            text_area.delete('1.0',END)
            text_area.insert(INSERT,"-------------------------------------------O U T P U T-------------------------------------------------")
            text_area.insert(INSERT,"\n")


            text_area.insert(END,f"{texts}")
            text_area.configure(state ='disabled')
            text_area.place(relx = 0.0001,rely = 0.0001,height = 600,width = 775)
            text_area.focus()
            extract(q_frame,a_frame_2,src_entr,image_list,element)

        except:

            a_lbl = Label(q_frame,text="Image Will Be Shown Here", fg = 'black',bg = 'gray',font = ('',20,'bold','underline'))
            a_lbl.place(relx = 0.2,rely = 0.4)
            messagebox.showerror("Unsupported file","File Not Supported !!!!!")
            return

home_screen() 
root.mainloop()
```
